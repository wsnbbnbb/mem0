```python
import ast

tree = ast.parse(s, mode="eval")
call = tree.body  # ast.Call
attrs = {}
for kw in call.keywords:
    attrs[kw.arg] = ast.literal_eval(kw.value)

vct = attrs["vector"]
payload = attrs["payload"]
# print(payload.keys())
# print(payload["user_id"])
# print(payload["text"])
text = payload["text"]
print(len(text))
```
